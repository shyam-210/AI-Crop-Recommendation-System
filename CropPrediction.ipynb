# Model Training

*We are going to make Classification models using this dataset*
    <br>*Classification model for classifying crop type to be grown based on the soil and atmospheric conditions

# Choosing the model for classification

```python
Classification_X=df1.drop(['CROP','CROP_PRICE'],axis=1)
Classification_Y=df1['CROP']
```

```python
Classification_X
```

```python
Classification_Y
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(42)
classxtrain,classxtest,classytrain,classytest = train_test_split(Classification_X,Classification_Y,test_size=0.2)
len(classxtrain)==len(classytrain)
```

```python
from sklearn.svm import LinearSVC

Class_model = LinearSVC(max_iter=1000,dual='auto')
Class_model.fit(classxtrain,classytrain)
Class_model.score(classxtest,classytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(estimator = Class_model , X=Classification_X , y=Classification_Y )
```

# checking using confusion matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(estimator=Class_model,X=Classification_X,y=Classification_Y)
```

```python
classytest
```

```python
from sklearn.metrics import classification_report
class_ypred = Class_model.predict(classxtest)
print(classification_report(classytest,class_ypred))
```

The Model is working perfectly in all evaluations
